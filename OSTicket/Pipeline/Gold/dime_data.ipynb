```python
%sql
CREATE OR REPLACE TABLE osticket.gold.dim_data AS
WITH limites AS (
    SELECT
        date_trunc('month', min(created)) AS min_data,
        date_trunc('month', max(created)) AS max_data
    FROM osticket.silver.thread_entry
),
datas AS (
    SELECT explode(
        sequence(
            min_data,
            max_data,
            interval 1 month
        )
    ) AS data
    FROM limites
)
SELECT
    data                           AS date,
    year(data)                     AS year,
    month(data)                    AS month_num,
    date_format(data, 'MMM')       AS month_name,
    date_format(data, 'MMMM')      AS month_full_name,
    date_format(data, 'yyyy-MM')   AS ano_month,
    year(data) * 100 + month(data) AS id_data
FROM datas
ORDER BY data;
```

```python
%sql
select * from osticket.gold.dim_data;
```
